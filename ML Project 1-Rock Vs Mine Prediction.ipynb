In the vast world of ocean travel, keeping ships and sailors safe is a top priority. This project focuses on the vital task of predicting whether objects underwater are rocks or mines.

In the context of the "Rock vs. Mine" prediction task, the sonar data likely includes information about the patterns of sound waves reflected off underwater objects. This data is then used to train machine learning models, allowing them to learn and predict whether the echoes correspond to rocks or mines based on the features extracted from the sonar signals.

### Segregate Dataset into X(Input/IndependentVariable) & Y(Output/DependentVariable)

```python

X = sonar_data.drop(columns=60, axis=1)
Y = sonar_data[60]
```

```python
print(X)
print(Y)
```

### Splitting Dataset into Train & Test

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.1, stratify=Y, random_state=1)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
print(X_train)
print(Y_train)
```

### Training

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
lr= LogisticRegression()
lr.fit(X_train,Y_train)
y_pred1=lr.predict(X_test)
```

### Predicting, whether the object is Mine or Rock

```python
input_data = (0.0307,0.0523,0.0653,0.0521,0.0611,0.0577,0.0665,0.0664,0.1460,0.2792,0.3877,0.4992,0.4981,0.4972,0.5607,0.7339,0.8230,0.9173,0.9975,0.9911,0.8240,0.6498,0.5980,0.4862,0.3150,0.1543,0.0989,0.0284,0.1008,0.2636,0.2694,0.2930,0.2925,0.3998,0.3660,0.3172,0.4609,0.4374,0.1820,0.3376,0.6202,0.4448,0.1863,0.1420,0.0589,0.0576,0.0672,0.0269,0.0245,0.0190,0.0063,0.0321,0.0189,0.0137,0.0277,0.0152,0.0052,0.0121,0.0124,0.0055)

# changing the input_data to a numpy array
input_data_as_numpy_array = np.asarray(input_data)

# reshape the np array as we are predicting for one instance
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

prediction = lr.predict(input_data_reshaped)
print(prediction)

if (prediction[0]=='R'):
  print('The object is a Rock')
else:
  print('The object is a mine')
```

### Evaluating Model - CONFUSION MATRIX

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(Y_test, y_pred1)

print("Confusion Matrix: ")
print(cm)

print("Accuracy of the Model: {0}%".format(accuracy_score(Y_test, y_pred1)*100))
```
